```python
#학번 부여 및 퀘스트 점수 관리하는 Aiffel 클래스를 만들기

import csv
import random

#이니셜라이저에서는 4개의 인자 (이름, 입학연도, 기수, 과정명)
class Aiffel():
    id_num = 0
    students= {'김건': 0, '김경민': 0, '김소연': 0, '김영신': 0, '박근수': 0, '박영수': 0, '박영준': 0, "서원영": 0, '신지만': 0, '심지안': 0, '이규상': 0, '이윤상': 0, '이진영': 0, '임현석': 0, '전요한': 0, '전재영': 0, '조대희':0}
    def __init__(self, name, admission_year, year, course):
        self.name = name
        self.admission_year = admission_year
        self.year = year
        self.course = course
        self.quest_point = 0
        self.score_sum = 0
        self.penalty_sum = 0
        self.id_num += 1
        self.student_id = generate_student_id(self)
#학번은 7자리로 이뤄져있고 학번 생성 규칙 YYNNCXX
#YY : 연도 마지막 두 자리 (ex. 2023년은 23) NN : 기수 (온라인 4기는 25기) C : 과정 종류 (리서치=1, 코어=2) XX : 가나다순 번호
    def generate_student_id(self):
        YY = str(self.admission_year)[-2:]
        NN = str(self.year).zfill(2)
        C = '1' if self.course_name == '리서치' else '2'
        sorted_students = sorted(self.students)

        XX = sorted_students.index(self.name) # 가나다순 번호는 초기값으로 XX를 사용
        student_id = YY + NN +str(C) + str(XX).zfill(2)
        return student_id


    @classmethod
    def get_id_num(cls):
        print(cls.id_num)


#퀘스트 점수 입력을 위한 score 메서드를 추가하세요. 점수는 +1점 부터 +3점까지만
    def score(self,score):
        if 1<= score <= 3:
            self.quest_point += score
            self.score_sum += score
        else:
            print('퀘스트 점수 입력은 1점부터 3점까지만 가능합니다.')

#퀘스트 점수 차감을 위한 penalty 메서드를 추가하세요. penalty로 인한 잔여 점수는 음수가 가능
    def penalty(self, penalty):
        self.quest_point -= penalty
        self.penalty_sum += penalty

# display_info() 메서드를 추가. 내용은 이름, 학번, 과정종류, 퀘스트 점수 총점, 퀘스트 입력 총점, 퀘스트 penalty 총합이 새로운 코어 명단(csv파일)로 출력
    def display_info(self, file_path):

        row1 = ['이름', '학번', '과정종류', '퀘스트 점수 총점', '퀘스트 입력 총점', '퀘스트 penalty 총점']
        row2 = [self.name, self.student_id, self.course,self.quest_point,self.score_sum,self.penalty_sum]

        with open(file_path, 'w', newline = '') as csvfile:
            writer = csv.writer(csvfile)
            writer.writerow(row1)
            writer.writerow(row2)



#코어 명단(딕셔너리)을 입력받아 학번을 부여하는 초기화 메서드를 작성해보세요



#중도포기자를 관리하기 위해 그루 정보를 삭제하는 메서드를 추가 메서드를 호출하면 명단이 갱신 김경민, 김영신 님을 삭제한후 명단 갱신
    @classmethod
    def delete(cls,del_name):
        if del_name in cls.studnets:
            del cls.students[del_name]


#Aiffel 클래스를 상속, 명단으로 2인, 3인, 4인으로 랜덤 그룹핑을 해주는 메서드가 추가된 Aiffel_Group 클래스 생성
#그룹 이름은 총 8개 입니다. (상생, 신뢰, 열정, 이끔, 성장, 가치, 공유, 확산)
class Aiffel_Group(Aiffel):
    Groups = ['상생', '신뢰', '열정', '이끔', '성장', '가치', '공유', '확산']

    def __init__(self):
        super().__init__()
        self.group_name = random.choice(Aiffel_Group.Groups)
        self.group_score = 0
#명단으로 2인, 3인, 4인으로 랜덤 그룹핑을 해주는 메서드가 추가된 Aiffel_Group 클래스 생성
    @staticmethod                                           #staticmethod 사용 이유 : 학생 명단이 바뀌면 안되기 때문
    def random_grouping(students):
        random.shuffle(students)
        group_size = [2,3,4]
        grouped_students = []
        idx = 0

        for size in group_size:
            group = students[idx:idx+size]
            grouped_students.append(group)
            idx += size
        return grouped_students

    def print_grouped_students(self):
        grouped_students = self.random_grouping(Aiffel.students)

        for i, group in enumerate(grouped_students, start=1):
            group_name = Aiffel_Group.Groups[i-1] + ' 그룹'
            print(f"{group_name}: {', '.join(group)}")

#Aiffel_Group 클래스에 group2guild_score 메서드를 추가,  그룹이름과 그룹점수 2개의 인자 그룹핑된 그루들이 소속된 그룹의 이름으로 점수를 받게 되면, 각자 소속된 길드의 점수로 연동
    def group2guild_score(self, group_name, group_score):

        self.group_name += group_score
        self.quest_point += group_score
        for students in group_name:

    @staticmethod                                           #staticmethod 사용 이유 : 학생 명단이 바뀌면 안되기 때문
    def random_guilding(students):
        random.shuffle(students)
        guilded_size = [2,3,4]
        idx = 0
        for size in guilded_size:
            guilded = students[idx:idx+size]
            guilded_students.append(guilded)
            idx += size
        return guilded_students

    def print_grouped_students(self):
        grouped_students = self.random_grouping(Aiffel.students)

        for i, group in enumerate(grouped_students, start=1):
            group_name = Aiffel_Group.Groups[i-1] + ' 그룹'
            print(f"{group_name}: {', '.join(group)}")
#Aiffel 클래스를 상속, 명단으로 “힌튼”, “알트만”, “응”, “하사비스” 네개의 길드로 그루들을 랜덤그룹핑해주는 Aiffel_Guild 클래스 생성
class Aiffel_Guild(Aiffel):
    Guilds = ['힌튼', '알트만','응','하사기스']
    def __init___(self):
        super().__init__()
        self.guild_name = random.choice(Aiffel_Guild.Guilds)
        self.guild_get = 0


#Aiffel_Guild 클래스에 길드별 점수를 입력하고 차감하는 guild_score 메서드를 추가, 메서드는 길드 이름과 점수를 인자로 받고 길드별 스코어를 출력
# guild_score 메서드로 점수가 입력되거나 차감되면, Aiffel 클래스에서 구현된 개별 그루의 score도 점수 영향
    def guild_score(self, score):
        self.score += score
        self.point += score


#Aiffel_Group 클래스와 Aiffel_Guild 클래스가 Aiffel클래스의 display_info() 메서드를 아래와 같이 오버라이딩
#guild_score(), group2guild_score() 메서드를 호출했을 시 기존 명단 +  길드별 스코어 총합이 보이도록 오버라이딩
```
